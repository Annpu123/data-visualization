```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('tips')
df
```

```python
df.head(1)
```

```python
figr=sns.FacetGrid(df)
figr.map (plt.scatter, 'total_bill','tip')
```

```python
figr=sns.FacetGrid(df)
figr.map(plt.bar, 'total_bill','tip')
```

```python
figr=sns.FacetGrid(df, col='sex')
figr.map (plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr=sns.FacetGrid(df, col='sex')
figr.map (plt.bar, 'total_bill','tip')
plt.show()
```

```python
figr=sns.FacetGrid(df, col='day')
figr.map (plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr=sns.FacetGrid(df, col='day',hue='sex')
figr.map (plt.scatter, 'total_bill','tip')
plt.show()
```

```python
figr=sns.FacetGrid(df, col='sex',hue='day')
figr.map (plt.scatter, 'total_bill','tip')
figr.add_legend()
plt.show()
```

```python
figr=sns.FacetGrid(df, col='day',hue='sex')
figr.map (plt.scatter, 'total_bill','tip')
figr.add_legend()
plt.show()
```

```python
figr=sns.FacetGrid(df, col='day',hue='sex',palette='cool')
figr.map (plt.scatter, 'total_bill','tip')
figr.add_legend()
plt.show()
```

# Pair plot

```python

df.head(1)
```

```python
plt.figure(figsize=(4,2))
sns.pairplot(df)
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='smoker')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df,vars=['size','tip'] ,hue='smoker')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df,vars=['size','tip'] ,hue_order=['Female','Male'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df,x_vars=['size','tip'],hue='sex' ,hue_order=['Female','Male'],y_vars=['total_bill','tip'],palette='hot')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent',kind='reg')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent',kind='kde')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent',kind='hist')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent',kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(df, hue='sex',hue_order=['Female','Male'],y_vars=['total_bill'],palette='Accent',kind='reg',markers=['D','*'])
plt.show()
```
